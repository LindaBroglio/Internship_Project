```python
%cd '/Users/linda/Desktop/Internship/Internship_Project/lung_data'

import scipy 
import sklearn
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import PowerTransformer
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.base import BaseEstimator, TransformerMixin
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import argparse
from LDA import get_x_y,binary_value
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
def from_xls_into_df(header):
    path, filename = "input_data", "input_file.xlsx"
    sheet_names = ['PET', 'CT']
    dfs = pd.read_excel(path + "/" + filename, sheet_name=sheet_names, header=header)
    return dfs['PET']
```

```python
class TypeToBinaryTransformer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X['Type'] = X['Type'].apply(lambda g: int(1) if g == "met" else int(0))
        return X
```

```python
df_PET = from_xls_into_df(1)
df_PET.drop('Patient_ID', axis=1, inplace=True)
x, y = get_x_y(df_PET, 'Type')
y['Type']=[1 if g == 'met' else 0 for g in y['Type']]
x_cols = x.columns
y_cols = y.columns
```

```python
features_pipe = Pipeline(steps=[
    ("missing steps", SimpleImputer(strategy='mean')),
    ("scaling", MinMaxScaler())
])
```

```python
target_pipe = Pipeline(steps=[
    #("type_to_binary", TypeToBinaryTransformer()), 
    ("missing_steps", SimpleImputer(strategy='most_frequent')),
    ("scaling", MinMaxScaler())
])
```

```python
ct = ColumnTransformer(
    [("features", features_pipe, list(x_cols)),
    ("target", target_pipe, list(y_cols))
    ])
```

```python
model = LinearDiscriminantAnalysis()
```

```python
model_pipe = Pipeline(steps=[
        ("preprocessing", features_pipe),
        ("model", model)]
)
```

```python
accuracies = []
for i in range(100):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=i)
    model_pipe.fit(x_train,y_train)
    y_pred = model_pipe.predict(x_test)
    accuracies.append(accuracy_score(y_test, y_pred))

mean_accuracy, std_accuracy = np.mean(accuracies), np.std(accuracies)
```
